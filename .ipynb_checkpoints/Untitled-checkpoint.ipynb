```python
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('./pics/Final_414 (1).jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image, cmap="gray")
```

```python
	
%load_ext cython
```

```python
%%cython -a
 
def threshold_slow(T, image):
    # grab the image dimensions
    h = image.shape[0]
    w = image.shape[1]
    
    # loop over the image, pixel by pixel
    for y in range(0, h):
        for x in range(0, w):
            # threshold the pixel
            image[y, x] = 255 if image[y, x] >= T else 0
            
    # return the thresholded image
    return image
```

```python
%timeit threshold_slow(5, image)
```

```python
1 loop, best of 3: 244 ms per loop
```

```python
# threshold our image to validate that it's working
image = threshold_slow(10, image)
plt.imshow(image, cmap="gray")
```

```python
%%cython -a
 
import cython
 
@cython.boundscheck(False)
cpdef unsigned char[:, :] threshold_fast(int T, unsigned char [:, :] image):
    # set the variable extension types
    cdef int x, y, w, h
    
    # grab the image dimensions
    h = image.shape[0]
    w = image.shape[1]
    
    # loop over the image
    for y in range(0, h):
        for x in range(0, w):
            # threshold the pixel
            image[y, x] = 255 if image[y, x] >= T else 0
    
    # return the thresholded image
    return image
```
